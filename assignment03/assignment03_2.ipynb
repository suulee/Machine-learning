```


## cost/loss function
cost=0.5*tf.reduce_mean(tf.square(hypothesis-Y))

## Minimize
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.01)
train=optimizer.minimize(cost)
#learning_rate=0.1
#gradient=tf.reduxce_mean((theta1*X+theta0 - Y))
#descent1= theta1-learning_rate*gradient
#update=theta1.assign(descent)

##launch the graph in a session
sess=tf.Session()

##initializes global variables in the graph
sess.run(tf.global_variables_initializer())

for step in range(6001):
    sess.run(train,feed_dict={X:x_data, Y:y_data})
    if step%200==0:
      print(step, sess.run(cost,feed_dict={X:x_data, Y:y_data}),sess.run(theta0,feed_dict={X:x_data, Y:y_data}),sess.run(theta1,feed_dict={X:x_data, Y:y_data}))
```
